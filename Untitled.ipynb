```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Electronic_sales_Sep2023-Sep2024.csv')
data.head()
```

```python
# 1 
data.groupby('Customer ID')['Payment Method'].nunique().sort_values(ascending = False)
```

```python
data.groupby('Customer ID')['Total Price'].sum().sort_values(ascending=False)
```

```python
data.groupby('Customer ID')['Add-on Total'].sum().sort_values(ascending = False)
```

```python
#2

by_shipping = data.groupby('Shipping Type')['Total Price'].sum().sort_values(ascending = False)
px.bar(by_shipping)
```

```python
by_product_type = data.groupby('Product Type')['Total Price'].sum().sort_values(ascending = False)
px.bar(by_product_type)
```

```python
by_add_on_and_date = data.groupby(['Add-ons Purchased', 'Purchase Date'])['Total Price'].sum().sort_values(ascending = False)
by_add_on_and_date.index = by_add_on_and_date.index.map(lambda x: x[0] + ' ' + x[1])
by_add_on_and_date

px.bar(by_add_on_and_date)
```

```python
data_copy = data.copy()
data_copy['Quarters'] = data_copy['Purchase Date'].map(lambda x: pd.Timestamp(x).quarter)
by_add_on_and_quarter = data_copy.groupby(['Add-ons Purchased', 'Quarters'])['Total Price'].sum().sort_values(ascending = False)
by_add_on_and_quarter.index = by_add_on_and_quarter.index.map(lambda x: x[0] + ' ' + str(x[1]) + 'q')

px.bar(by_add_on_and_quarter)
```
